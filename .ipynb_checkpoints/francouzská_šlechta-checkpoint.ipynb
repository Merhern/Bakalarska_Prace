```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {"Období": ["1650-99", "1700-49", "1750-99", "1800-49"],
        "Průměrný počet dětí v manželství": [6.15, 2.79, 2, 3.72]}
df = pd.DataFrame(data)
```

```python
sns.catplot(data=df, y="Průměrný počet dětí v manželství", x="Období", kind="bar", color="blue")

plt.title("Úpadek porodnosti francouzské vyšší šlechty")
plt.xlabel("Období francouzské historie")
plt.ylabel("Průměrný počet dětí v manželství vyšší šlechty")
plt.show()
```

```python
data = {"Období": ["1650-99", "1700-49", "1750-99", "1800-49", "1650-99", "1700-49", "1750-99", "1800-49"],
       "Průměrný věk při prvním manželství": [25.5, 23.6, 21.3, 28.2, 20, 19.4, 18.3, 21.3],
        "Pohlaví": ["Muž", "Muž", "Muž", "Muž", "Žena", "Žena", "Žena", "Žena"]}
df = pd.DataFrame(data)

sns.catplot(data=df, y="Průměrný věk při prvním manželství", x="Období", hue="Pohlaví", kind="bar")
plt.title("Průměrný věk při prvním manželství lidí narozených do francouzské vyšší šlechty")
plt.ylabel("Průměrný věk při prvním manželství")
plt.xlabel("Období francouzské historie")
plt.show()
```

```python
sns.catplot(data=df, y="Průměrný věk při prvním manželství", x="Období", hue="Pohlaví", kind="point")
plt.show()
```
