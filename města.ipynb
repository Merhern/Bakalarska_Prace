```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("OD_UZE01_2024051613281094.CSV")

hodnoty = ["Hradec Králové", "Pardubice", "Liberec", "Ústí nad Labem", "Olomouc"]

df = df[df['vuzemi_txt'].isin(hodnoty)]

print(df['vuzemi_txt'].value_counts())
```

```python
df = df[ (df["stapro_txt"] == "Počet obyvatel s trvalým nebo dlouhodobým pobytem") 
    & ( df["vuzemi_cis"] == 65) & ( df["velskup_txt"] == "celkem")]
df2 = df.groupby(["vuzemi_txt", "rok"])["hodnota"].sum().reset_index()

sns.relplot(
    data=df2, x="rok", y='hodnota', hue="vuzemi_txt", kind="line", errorbar=None
)

plt.show()
```

```python
sns.relplot(
    data=df2, x="rok", y='hodnota', hue="vuzemi_txt", kind="scatter"
)

plt.show()
```
